```python
import pandas as pd
```

```python
# create a movie Dataset
# Step 1: Create the Movie Dataset
data = {
    'Movie Name': ['Inception', 'Avatar', 'Titanic', 'The Dark Knight', 'Interstellar', 'The Lion King', 'The Avengers', 'Shrek', 'Avengers: Endgame', 'Frozen'],
    'Director': ['Christopher Nolan', 'James Cameron', 'James Cameron', 'Christopher Nolan', 'Christopher Nolan', 'Jon Favreau', 'Joss Whedon', 'Andrew Adamson', 'Anthony Russo, Joe Russo', 'Chris Buck, Jennifer Lee'],
    'Release Year': [2010, 2009, 1997, 2008, 2014, 2019, 2012, 2001, 2019, 2013],
    'Genre': ['Sci-Fi', 'Sci-Fi', 'Romance', 'Action', 'Sci-Fi', 'Animation', 'Action', 'Animation', 'Action', 'Animation'],
    'Rating': [8.8, 7.8, 7.8, 9.0, 8.6, 8.5, 8.0, 7.9, 8.4, 7.5],
    'Budget (in millions)': [160, 237, 200, 185, 165, 260, 220, 60, 356, 150],
    'Box Office Revenue (in millions)': [829, 2790, 2187, 1005, 677, 1650, 1519, 484, 2798, 1270]
}
```

```python
# create a dataframe

df = pd.DataFrame(data)
```

```python
# display first few rows

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
high_rated_movies = df[df['Rating'] > 8.0]
print(high_rated_movies)
```

```python
sorted_by_year = df.sort_values(by = 'Release Year', ascending = False)
print(sorted_by_year)
```

```python
high_revenue_movies = df[df['Box Office Revenue (in millions)'] > 500]
print(high_revenue_movies)
```

```python
sorted_by_rev_value = df.sort_values(by = 'Box Office Revenue (in millions)', ascending = False)
print(sorted_by_rev_value)
```

```python
average_rating = df['Rating'].mean()
print(average_rating)
```

```python
total_budget = df['Budget (in millions)'].sum()
print(total_budget)
```

```python
highest_grossing_movie = df.sort_values(by = 'Box Office Revenue (in millions)', ascending = False).iloc[0]
print(highest_grossing_movie)
```

```python
# grouping

avg_rating_by_genre = df.groupby('Genre')['Rating'].mean()
print(avg_rating_by_genre)
```

```python
tot_revenue_by_director = df.groupby('Director')['Box Office Revenue (in millions)'].sum()
print(tot_revenue_by_director)
```

```python
# calculate profits

df['Profit (in millions)'] = df['Box Office Revenue (in millions)'] - df['Budget (in millions)']
print("Dataset with Profit column:")
print(df[["Movie Name", "Profit (in millions)"]], "\n")


highest_profit_movie = df.loc[df['Profit (in millions)'].idxmax()]
print(highest_profit_movie)
```
